In the literature that was assigned we are given that the model is based on the Black-Scholes setting meaning that we have the following equations to describe the market
\begin{align*}
    \begin{cases}
    dB_t = rB(t) dt, \\
    dS(t) = \alpha(t,S(t)) S(t) dt + \sigma(t,S(t)) S(t) dW(t).
    \end{cases}
\end{align*}
Here $B_t$ denotes the the risk free asset and $S_t$ is the stock price process. Further $r$ is the short rate of interest. So far the stock price is described with $\mathbb{P}$-measure, using the risk free asset as a numeraire we obtain the following $\mathbb{Q}$-measure where the stock price process is given by
\begin{align*}
    dS(t) = rS(t) dt + \sigma(t,S(t))  S(t) dW^Q(t)
\end{align*}
This implies that $\sigma(t,S(t))$, which is denoted the local volatility, is of importance.

### Task 1

Describe the stochastic model for the stock price in a local volatility model and then
explicitly derive the Dupire formula by going through the steps I-III as outlined in
the document “The local volatility surface”.

#### Step 1

We begin by letting $max(X(t)-x,0) = f(X_t)$. Then as given in the Eq.(4) in the notes we have
\begin{equation*}
    f(X_t) = f(X_0) + \int_{0}^{t} f'(X_s) dX_s + \frac{1}{2} \int_{-\infty}^{\infty} L_t^y(X) d\mu(y).
\end{equation*}
Now the differentation of Eq.(4) gives
\begin{equation*}
    df(X_t) = f'(X_t) dX_t + \frac{1}{2} dL_t^x(X).
\end{equation*}
As given in the derivation of the formula, integration by parts is needed here and from the lecture notes it has the following structure.
\begin{align*}
    d(XY)_t = X_tdY_t + Y_tdX_t + d<X,Y>_t
\end{align*}
Let $Y_t = e^{-rt}$ and the following is obtained
\begin{align*}
    d(f(X_t)e^{-rt})_t &= f(X_t)d(e^{-rt})_t + e^{-rt}d(f(X_t))  + df(X_t)de^{-rt} \\ &= -rf(x_t) e^{-rt}dt +e^{-rt}(f'(X_t) dX_t + \frac{1}{2} dL_t^x(X))
    \\  &= -rf(x_t) e^{-rt}dt +e^{-rt}f'(X_t) dX_t + e^{-rt}\frac{1}{2} dL_t^x(X))
\end{align*}
We note here that from the exercise lessons $df(X_t)de^{-rt} = 0$ due to the condition of $dt$ and $dW$, as the only values that are nonzero are the multiplications of $dW \cdot dW = dt$. Integration from $0$ to $T$ in combination of substitution of $X = S$ and $x = K$ gives
\begin{align*}
    f(X_T)e^{-rT} - f(X_0)e^{-r0} = -r \int_{0}^T f(X_u) e^{-ru}du + \int_{0}^T e^{-ru}f'(X_u) dX_u + \int_{0}^T e^{-ru}\frac{1}{2} dL_u^x(X))
\end{align*}
\begin{align*}
    f(S_T)e^{-rT}  &=  f(S_0) -r \int_{0}^T f(S_u) e^{-ru}du + \int_{0}^T e^{-ru}f'(S_u) dS_u + \frac{1}{2} \int_{0}^T e^{-ru} dL_u^K(S)) \\
    &= \max(S(0)-K,0) -r \int_{0}^T \max(S(u)-K,0) e^{-ru}du + \int_{0}^T e^{-ru}\mathbf{1}_{(S(u)>K)} dS_u \\ & + \int_{0}^T e^{-ru}\frac{1}{2} dL_u^K(S))
\end{align*}

#### Step 2

We have two equations that are needed to move further which are

\begin{align*}
    c(T,K) = e^{-rt}\mathbb{E}^Q[max(S(T)-K,0)],
\end{align*}
and 
\begin{align*}
    \mathbb{E}[dL_t^x(X)] = \sigma^2(t,x)x^2p(t,x)dt.
\end{align*}
For our case
\begin{align*}
    \mathbb{E}[dL_u^K(S)] = \sigma^2(u,K)K^2p(u,K)du.
\end{align*}
Moreover, taking the expectation of the result in the previous section we obtain
\begin{align*}
    c(T,K) & =\max(S(0)-K,0) +  \mathbb{E}[ -r \int_{0}^T \max(S(u)-K,0) e^{-ru}du + \int_{0}^T e^{-ru}\mathbf{1}_{(S(u)>K)} dS_u ] \\ &+ \frac{1}{2} \int_{0}^T e^{-ru} K^2 \sigma^2(u,K)q(u,K)du.
\end{align*}
Here we utilize that the fact that from a previous course we learned that the expectation can be moved inside the integral $\mathbb{E}[\int_{0}^T e^{-ru}\frac{1}{2} dL_u^K(S))] = \int_{0}^T e^{-ru}\frac{1}{2} \mathbb{E}[dL_u^K(S))] $.
Further it is noted that $\max(S(u)-K,0)$ can be rewritten as 
\begin{align*}
    \max(S(u)-K,0) = \mathbf{1}_{(S(u)>K)}(S(u)-K),
\end{align*}
and in the assignment we have been given
\begin{align*}
    dS(t) = rS(t)dt + \sigma S(t) dW(t).
\end{align*}
To find the expectation is necessary of us to rewrite the following
%first integral
\begin{align*}
      -r \int_{0}^T \mathbf{1}_{(S(u)>K)}(S(u)-K)e^{-ru}du = -r \int_{0}^T \mathbf{1}_{(S(u)>K)}S(u)e^{-ru}du +r \int_{0}^T \mathbf{1}_{(S(u)>K)}Ke^{-ru}du, 
\end{align*}
and 
%second integral
\begin{align*}
     \int_{0}^T e^{-ru}\mathbf{1}_{(S(u)>K)} dS_u = \int_{0}^T e^{-ru}\mathbf{1}_{(S(u)>K)} (rS(u)du + \sigma S(u) dW(u)).
\end{align*}
Under the expectation the final summation is 
\begin{align*}
    \mathbb{E}[-r \int_{0}^T \mathbf{1}_{(S(u)>K)}S(u)e^{-ru}du +r \int_{0}^T \mathbf{1}_{(S(u)>K)}Ke^{-ru}du
      + \int_{0}^T e^{-ru}\mathbf{1}_{(S(u)>K)} (rS(u)du + \sigma S(u) dW(u))].
\end{align*}
We note here that a stochastic integral under the expectation is equal to zero, hence $\int_0^T ... dW_u = 0$, which is our final term in the expectation above, we are then left with
\begin{align*}
    \mathbb{E}[-r \int_{0}^T \mathbf{1}_{(S(u)>K)}S(u)e^{-ru}du +r \int_{0}^T \mathbf{1}_{(S(u)>K)}Ke^{-ru}du
      + r\int_{0}^T e^{-ru}\mathbf{1}_{(S(u)>K)} S(u)du].
\end{align*}
The result is that two of integrals cancel each other out. Hence the expectation that is left in our equation above is equal to 
\begin{align*}
   \mathbb{E}[ -r \int_{0}^T \mathbf{1}_{(S(u)>K)}(S(u)-K)e^{-ru}du + \int_{0}^T e^{-ru}\mathbf{1}_{(S(u)>K)} dS_u ] &= \mathbb{E}[r\int_{0}^T\mathbf{1}_{(S(u)>K)}Ke^{-ru}du] \\ &=  rK\int_{0}^T e^{-ru}Q(S(u) > K) du.
\end{align*}
We arrive at
\begin{align*}
    c(T,K) = \max(S(0)-K,0)+ rK\int_{0}^T e^{-ru}Q(S(u) > K) du + \frac{1}{2} \int_{0}^T e^{-ru} K^2 \sigma^2(u,K)q(u,K)du.
\end{align*}

#### Step 3

Using Eq.(5) in the notes we have
% More information can be found on the indicator function Wikipedia, at the expectation!
% https://en.wikipedia.org/wiki/Indicator_function
% https://www.csie.ntu.edu.tw/~d00922011/python/cases/LocalVol/DUPIRE_FORMULA.PDF
% Portfolio course book
\begin{align*}
    \frac{\partial c}{\partial K} = e^{-rT}\frac{\partial \mathbb{E}^Q[\max(S(T)-K,0)]}{\partial K}  &= e^{-rT}\frac{\partial \mathbb{E}^Q[\mathbf{1}_{S(T)>K}(S(T)-K)]}{\partial K} \\
    &= e^{-rT}\frac{\partial (S(T)-K)  Q(S(T)>K))}{\partial K} \\ &=
    -e^{-rT}Q(S(T)>K)
\end{align*}
Further we have (from probability theory we know that $ \int_{K}^{\infty} q(T,x) dx  = (-\Phi +1 )$)
\begin{align*}
    \frac{\partial^2 c}{\partial K^2} = -e^{-rT}\frac{\partial Q(S(T)>K)}{\partial K} = -e^{-rT} \frac{\partial}{\partial K} \int_{K}^{\infty} q(T,x) dx = -e^{-rT} \frac{\partial}{\partial K} (-\Phi +1 ) = e^{-rT} q(T,K).
\end{align*}
At last differentiation by T of the main equation gives
\begin{align*}
    \frac{\partial c}{\partial T} &= rK e^{-rT}Q(S(T) > K) +  \frac{1}{2} e^{-rT} K^2 \sigma^2(T,K)q(T,K) \\ &= -rK \frac{\partial c}{\partial K}   +  \frac{1}{2} K^2 \sigma^2(T,K)\frac{\partial^2 c}{\partial K^2}.
\end{align*}
Which if we rearrange we have
\begin{align*}
    \sigma(T,K) = \sqrt{2\frac{\frac{\partial c}{\partial T}(T,K) + rK \frac{\partial c}{\partial T} (T,K)}{K^2\frac{\partial^2 c}{\partial K^2}(T,K)}},
\end{align*}
which is Dupire's formula.

### Task 2

Use call option prices from the two csv-files found at the homepage of the course
to estimate two local volatility surfaces using Dupire's formula. Each row in the 
csv-files consists of data on the form

Time to maturity, Strike price, Call option price.

Set r = 0 and approximate the derivatives in the Dupire formula using finite differences.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
from mpl_toolkits.mplot3d import Axes3D
```

```python
options_1 = pd.read_csv('New_option_prices_1.csv')
options_2 =pd.read_csv('New_option_prices_2.csv')
#For New_option_prices_1 the spot price is approximately 1593; for New_option_prices_2 the spot price is approximately 1624
```

```python
#options_1['dt_backwards'] = options_1.groupby('Strike').apply(lambda x: x['Price'].diff(-1)/x['Maturity'].diff(-1)).reset_index(level=0)[0]
options_1['dt_forwards'] = options_1.groupby('StrikePrice').apply(lambda x: x['CallOptionPrice'].diff(-1)/x['TimeToMaturity'].diff(-1)).reset_index(level=0)[0]
options_1['dK2'] = options_1[['TimeToMaturity', 'StrikePrice','CallOptionPrice']].groupby('TimeToMaturity').apply(lambda x: (x['CallOptionPrice'].shift(1) - 2*x['CallOptionPrice'] + x['CallOptionPrice'].shift(-1))/100).reset_index(level=0)['CallOptionPrice']
#h^2 = 100. 
options_1['local_vol'] = np.sqrt(2*options_1['dt_forwards']/(options_1['StrikePrice']**2 *options_1['dK2']))
```

```python
#%matplotlib widget

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

# Make data.
X = options_1['TimeToMaturity'].unique()
Y = options_1['StrikePrice'].unique()
X, Y = np.meshgrid(X, Y)
Z = []
for i in range(5):
    Z.append(options_1['local_vol'].iloc[i*21:(i+1)*21].values)
Z = np.transpose(np.array(Z))

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

# Customize the z axis.
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.title('local volatility surface, options_1')
plt.ylabel('Strike Price')
plt.xlabel('Time To Maturity')
fig.set_size_inches(18.5, 10.5)
plt.show()
```
